## 9.1 주성분을 사용해 특성 줄이기
* 일련의 특성이 주어질 때 분산 유지하면서 특성 수 줄이기
* 사이킷런의 PCA 사용

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import datasets

digits = datasets.load_digits()

# 특성 행렬 표준화
features = StandardScaler().fit_transform(digits.data)

# 99% 분산 유지하도록 PCA 클래스 객체 생성
pca = PCA(n_components=0.99, whiten=True)

# PCA 수행
features_pca = pca.fit_transform(features)

# 결과 확인
print("원본 특성 개수:", features.shape[1])
print("줄어든 특성 개수:", features_pca.shape[1])
```

* PCA가 10개의 차원을 줄이면서 특성 행렬의 정보(분산)을 99% 유지했다.

* 화이트닝(whitening) : 주성분에 투영된 특성의 스케일을 맞춘다.
    * PCA 평균이 0으로 맞추므로 화이트닝 대신 나중에 표준화도 가능하다.

```python
import matplotlib.pyplot as plt

plt.scatter(features_pca[:, 0], features_pca[:, 1])
plt.show()
```

```python
# whitening 적용하지 않으면 평균은 0이나 스케일은 맞춰지지않는다.
pca_nowhiten = PCA(n_components=0.99)
features_nowhiten = pca_nowhiten.fit_transform(features)
plt.scatter(features_nowhiten[:, 0], features_nowhiten[:, 1])
plt.show()
```

```python
# PCA 찾은 주성분이 components_ 속성에 저장되어 있다.
pca_nowhiten.components_.shape
```

```python
# 특성 행렬을 주성분에 투영하려면 components_ 배열을 전치하여 dot product 내적하면 된다.
# 넘파이 allclose 함수로 앞에 구한 features_nowhiten 배열과 동일한지 확인한다.
import numpy as np

np.allclose(features_nowhiten, np.dot(features, pca_nowhiten.components_.T))
```

* 전체 주성분의 설명된 분산에서 유지되는 분산 양이 크게 늘지 않은 지점을 찾을 수 있다.

```python
pca = PCA(whiten=True).fit(features)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.show()
```

* 여기서는 약 30개 주성분으로도 80% 이상의 분산을 유지하며 그 이후로 증가 추세가 꺾인다.(분산 누적)
* 샘플의 자유도 : 샘플 수 - 1
* 사전에 표준화 전처리 하면 다르다. 올바른 분산은 원본 데이터 그대로 PCA 주입한다.

```python
# 표준화하지 않은 원본 데이터를 사용한다.
pca.fit(digits.data)
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.show()
```

## 9.2 선형적으로 구분되지 않은 데이터의 차원 축소하기


* 커널 트릭 사용하는 주성분 분석의 확장을 사용해 비선형 차원 축소 수행 가능하다.

```python
from sklearn.decomposition import KernelPCA
from sklearn.datasets import make_circles

# 선형 구분 안되는 데이터 생성
features,_ = make_circles(n_samples=1000, random_state=1, noise=0.1, factor=0.1)
```

```python
kpca = KernelPCA(kernel="rbf", gamma=15, n_components=1)
features_kpca = kpca.fit_transform(features)

print("원본 특성 개수:", features.shape[1])
print("차원 축소된 특성 개수:", features_kpca.shape[1])
```

```python
kpca.components_
```

* 커널 PCA : 고차원 공간에서 주성분으로 투영된 결과 반환(실제 고차원 공간 변환하는 게 아니므로 주성분을 얻을 수 없다.)
* 즉, components를 반환하지 않는다.

## 9.3 클래스 분리를 최대화하여 특성 줄이기

* 분류 모델에 사용될 특성 줄인다.
* 선형 판별 분석(Linear Discriminant Analysis, LDA) : 클래스 최대한 분리하는 성분 축으로 특성 투영

```python
from sklearn import datasets
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

iris = datasets.load_iris()
features = iris.data
target = iris.target

# LDA 객체 만들고 실행하여 특성을 변환한다.
lda = LinearDiscriminantAnalysis(n_components=1)
features_lda = lda.fit(features, target).transform(features)

print("원본 특성 개수:", features.shape[1])
print("줄어든 특성 개수:", features_lda.shape[1])
```

* explained_variance_ratio_ : 각 성분이 설명하는 분산 양 확인

```python
lda.explained_variance_ratio_
```

```python
# LDA를 만들고 실행한다.
lda = LinearDiscriminantAnalysis(n_components=None)
features_lda = lda.fit(features, target)

# 설명된 분산의 비율이 담긴 배열 저장
lda_var_ratios = lda.explained_variance_ratio_

# 함수 제작
def select_n_components(var_ratio, goal_var : float) -> int:
    total_variance = 0.0 # 설명된 분산 초기값 지정
    
    n_components = 0 # 특성 개수 초깃값 지정
    
    for explained_variance in var_ratio: # 각 특성의 설명된 분산 순회
        
        total_variance += explained_variance # 설명된 분산 값 누적
        
        n_components +=1 # 성분 개수 카운트
        
        if total_variance >= goal_var: # 설명된 분산이 목표치 도달하면, 반복 종료
            break
            
    return n_components # 성분 개수 반환
```

```python
select_n_components(lda_var_ratios, 0.95)
```

## 9.4 행렬 분해를 사용하여 특성 줄이기
* 음수 아닌 특성 행렬이 있다면, 차원 축소에 비음수 행렬 분해(NMF)을 쓴다.

```python
from sklearn.decomposition import NMF
from sklearn import datasets

digits = datasets.load_digits()

features = digits.data

nmf = NMF(n_components=10, random_state=1)
features_nmf = nmf.fit_transform(features)

# 결과 출력
print("원본 특성 개수:", features.shape[1])
print("줄어든 특성 개수:", features_nmf.shape[1])
```

```python
nmf.components_.shape
```

```python
np.all(nmf.components_ >= 0)
```

```python
np.mean(features - np.dot(features_nmf, nmf.components_))
```

$$\mathbf{V} \approx \mathbf{W}\mathbf{H}$$

```python
# NMF 클래스의 solver 매개변수 기본값은 cd로 좌표 하강법 사용
# solver : 곱셈 업데이트 알고리즘 mu 옵션 추가
nmf_mu = NMF(n_components=10, solver='mu', random_state=1)
features_nmf_mu = nmf_mu.fit_transform(features)

np.mean(features - np.dot(features_nmf_mu, nmf_mu.components_))
```

## 9.5 희소한 데이터의 특성 줄이기
* TSVD(truncated singular value decomposition) 사용

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import TruncatedSVD
from scipy.sparse import csr_matrix
from sklearn import datasets
import numpy as np

digits = datasets.load_digits()

features = StandardScaler().fit_transform(digits.data)

features_sparse = csr_matrix(features)
```

```python
# TSVD 객체
tsvd = TruncatedSVD(n_components=10)

features_sparse_tsvd = tsvd.fit(features_sparse).transform(features_sparse)

print("원본 특성 개수:", features_sparse.shape[1])
print("줄어든 특성 개수:", features_sparse_tsvd.shape[1])
```

```python
# 처음 세 개의 성분이 설명하는 분산의 비율 합
tsvd.explained_variance_ratio_[0:3].sum()
```

```python
# 특성 개수보다 하나 작은 TSVD 제작 후 실행
tsvd = TruncatedSVD(n_components=features_sparse.shape[1]-1)
features_tsvd = tsvd.fit(features)

# 설명된 분산을 리스트에 저장
tsvd_var_ratios = tsvd.explained_variance_ratio_
```

```python
# 함수 제작
def select_n_components(var_ratio, goal_var):
    total_variance = 0.0
    
    n_components = 0
    
    for explained_variance in var_ratio:
        
        total_variance += explained_variance
        
        n_components += 1
        
        if total_variance >= goal_var:
            
            break
            
            
    # 성분 개수 반환
    return n_components
```

```python
# 함수 실행
select_n_components(tsvd_var_ratios, 0.95)
```

* 자연어 처리에서는 TSVD = 잠재 의미 분석 LSA

```python
features = digits.data - np.mean(digits.data, axis=0)

pca = PCA(n_components=40, random_state=1)
features_pca = pca.fit_transform(features)

tsvd = TruncatedSVD(n_components=40, random_state=1)
features_tsvd = tsvd.fit_transform(features)

np.max(np.abs(features_pca - features_tsvd))
```

* PCA 클래스의 svd_solver : auto
    * '<=' 500 : SVD 분해 사용하며 full이 됩니다.
    * '>' 500 : 랜덤 SVD 사용하는 randomized 된다.
* TruncatedSVD : algorithm 매개변수 기본값은 랜덤 SVD(randomized)
