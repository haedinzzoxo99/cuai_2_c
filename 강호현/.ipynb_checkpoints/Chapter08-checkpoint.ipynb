## 8.1 이미지 로드하기
* OpenCV :: imread 사용

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

image = cv2.imread("images/plane.jpg", cv2.IMREAD_GRAYSCALE)
```

```python
plt.imshow(image, cmap="gray"), plt.axis("off")
plt.show()
```

```python
type(image)
```

```python
# 이미지를 행렬로 변환했다.
image
```

```python
image.shape
```

* 해상도가 3600 * 2270으로 차원과 동일함을 알 수 있다.

```python
image[0,0] # 이미지 가장 우측 상단 픽셀의 강도
```

* 흑백에서 강도는 검정색(0)부터 흰색(255)로 구성되어 있다.
* 행렬의 각 원소는 파랑, 초록, 빨강 세 개의 값을 가질 수 있다.

```python
# 컬러 추가
image_bgr = cv2.imread("images/plane.jpg", cv2.IMREAD_COLOR)
```

```python
image_bgr[0,0]
```

```python
# 기본적으로 opencv :: BGR 사용하고, 나머지는 RGB 쓴다. 
# 그래서 RGB로 변환해야 한다.
image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb), plt.axis("off")
plt.show()
```
